More information on [GitHub](https://github.com/jupyter-widgets/ipyleaflet) and [readthedocs](https://ipyleaflet.readthedocs.io/en/latest/usage.html)

```python
import labInfo2JSON
JSONLABS = '/Users/otto/GitLab/inframsl/labs.json'
MSL_Labs = labInfo2JSON.loadJSONFromFile(JSONLABS)
```

```python
# show JSON (MSL_LABS)
prettyOutput = labInfo2JSON.getJSONPrettyDumpAsString(MSL_Labs)
print(prettyOutput)
```

```python
from ipyleaflet import *
from ipywidgets import Label
import geocoder # currently not used
center = (52.312, 7.647)

m = Map(
    center=(52.312, 7.647),
    zoom=4
)
display(m)

labelMouseOver = Label()
display(labelMouseOver)

def handle_interaction(**kwargs):
    if kwargs.get('type') == 'mousemove':
        labelMouseOver.value = str(kwargs.get('coordinates'))

m.on_interaction(handle_interaction)
```

```python
for lab in MSL_Labs['infrastructures']:
    lab_location = (lab['facility']['gps']['gps_lat'],lab['facility']['gps']['gps_lon'])
    #print(location)
    tooltip_text = lab['facility']['facility_name'] + '\n' + lab['facility']['address']['city']
    marker = Marker(location=lab_location, draggable=False, title=tooltip_text)
    m.add_layer(marker);
```
